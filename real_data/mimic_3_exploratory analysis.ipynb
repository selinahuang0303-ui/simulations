```python
import pandas as pd

df = pd.read_csv("../data/mimic3_sepsis_data.csv")
print(df.columns.tolist())
```

```python
for col in df.columns:
    unique = sorted(df[col].dropna().unique())

    if len(unique) <= 5:
        print(f"{col}: {unique}")
```

```python
binary_cols = [c for c in df.columns if set(df[c].dropna().unique()).issubset({0, 1})]

for col in binary_cols:
    print(f"\n{col}")
    print(df[col].value_counts())
```

| Variable        | Type       | Meaning                            | Confidence                              |
| --------------- | ---------- | ---------------------------------- | --------------------------------------- |
| `mechvent`      | Binary     | 1 = mechanically ventilated        | High (verified from FiO₂ and P/F ratio) |
| `vaso_input`    | Ordinal    | Vasopressor treatment level (0–4)  | High (verified from median dose)        |
| `iv_input`      | Ordinal    | IV fluid treatment level (0–4)     | High (verified from input volume)       |
| `mortality_90d` | Binary     | 1 = died within 90 days            | High                                    |
| `died_in_hosp`  | Binary     | 1 = died during hospitalization    | High                                    |
| `gender`        | Binary     | 0/1, coding unknown                | Low                                     |
| `reward`        | Continuous | RL reward? not a clinical variable | High                                    |
