```python
import sys

sys.path.append("../src")

from generate_predictors import generate_predictors
from correlation import introduce_correlation
from sklearn.model_selection import train_test_split
import numpy as np
```

```python


np.random.seed(42)


X1_CONFIG = {
    "size": 500,
    "dimension": 5,
    "variable_type": "continuous",
    "distribution": "normal",
    "distribution_params": {

        "mean": 0,

        "std": 1

    }
}

X2_CONFIG = {
    "size": 500,
    "dimension": 500,
    "variable_type": "continuous",
    "distribution": "normal",
    "distribution_params": {

        "mean": 2,

        "std": 5

    }
}

X1 = generate_predictors(X1_CONFIG)
X2 = generate_predictors(X2_CONFIG)
```

```python
print(X1.shape)
print(X2.shape)
```

```python
correlation_config = {
    "method": "linear",
    "strength": 0.6
}

X1_corr, X2_corr = introduce_correlation(X1, X2, correlation_config)
```

```python
import numpy as np

corr = np.corrcoef(X1_corr[:, 4], X2_corr[:, 4])[0, 1]

print(corr)
```

```python
print(X2_corr[:,0].mean())
print(X2_corr[:,0].std())
```

```python
X = np.hstack((X1, X2))

noise = np.random.normal(
    loc=0,
    scale=1,
    size=X.shape[0]
)

Y = (
    3 * X1[:, 0]
    + 2 * X2[:, 1]
    + noise
)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    test_size=0.2,
    random_state=42
)
```

```python
from models.lasso import fit_lasso

lasso_model = fit_lasso(X_train, Y_train, X_test, Y_test, alpha=0.5)
```

```python
print("MSE:", lasso_model["mse"])
print("Runtime:", lasso_model["runtime"])
print("Number selected:", lasso_model["n_selected"])
print("Selected variables:", lasso_model["selected_variables"][:10])
print("First 10 coefficients:", lasso_model["coefficients"][:10])
```

```python
from models.ridge import fit_ridge

ridge_model = fit_ridge(X_train, Y_train, X_test, Y_test, alpha=1000.0)
```

```python
print("MSE:", ridge_model["mse"])
print("Runtime:", ridge_model["runtime"])
print("Number selected:", ridge_model["n_selected"])
print("Selected variables:", ridge_model["selected_variables"][:10])
print("First 10 coefficients:", ridge_model["coefficients"][:10])
```

```python
from models.random_forest import fit_random_forest

rf_model = fit_random_forest(X_train, Y_train, X_test, Y_test, n_estimators=100, max_depth=5)
```

```python
print("MSE:", rf_model["mse"])
print("Runtime:", rf_model["runtime"])
print("Number selected:", rf_model["n_selected"])
print("Selected variables:", rf_model["selected_variables"][:10])
print("First 10 coefficients:", rf_model["coefficients"][:10])
```
